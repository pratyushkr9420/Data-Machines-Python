## **Sports Analytics**

Sports analytics is applying statistics, datascience, machine learning to both
on-field and off-field sports related data. The main goal is to somehow gain a competitive advantage (more wins, more fans, morerevenue, more profit) by learning insightsabout a team or its competition.

The field was popularized by the
[movie](https://www.imdb.com/title/tt1210166/)

## **Baseball & MLB**

Probably the most analyzed and data rich sport is baseball. The game of
baseball
is  dissected for offensive (hitting, base running), pitching, and defense (e.g. positioning playersbased on the situation).

MLB (Major League Baseball) is the American professional baseball organization. It has a rich history of providing information on each game and has a strong community in
[information and research](https://sabr.org/sabermetrics). One of the main

Statcast data is collected in all the MLB stadiums. The first generation

###**Pitch Movement**

As the ball is moving from the pitcher to the catcher, it's flight is
influenced (in addition to gravity) by it's spin, velocity, and the angle
from which the ball leaves the pitcher's hand. Ball

how much the ball moves due the pitcher's action. For many of baseball's

### **A Side Note on Pitch Movement (for the curious)**

On the baseball savant page, it says 'Justin Verlander's horizontal & vertical movement with gravity in inches'. The issue is that the
pfx_z
values in the downloaded data are 'InducedVertical Break' values. That is, values with the effect of gravity
removed





```
 Sport's Log: There is no clear definition of the pfx_z field given to confirm the above other than visual inspection of the data. 
MLB's definition of pitch movement (http://m.mlb.com/glossary/statcast/pitch-movement) does not align with its own definitions 
(https://baseballsavant.mlb.com/csv-docs) or with Trackman's (https://trackman.zendesk.com/hc/en-us/articles/115002776647-Radar-Measurement-Glossary-of-Terms).
However, the good people at Trackman did confirm the following: 
Statcast's pfxz represents the same measurement as Trackman's Induced Vertical Break, not Trackman's pfxz.
```



A few quick definitions:


*   **Vertical Break**
: the actual pitch compared to a theoretical pitch with no gravity.
*   **Induced Vertical Break**
: the actual pitch compared to a theoretical pitch
affected bygravity.


Basically, induced vertical break allows one to understand how much the
pitcher is actually influencing the movement of the ball.

A reference to a 'rising' pitch, just means the ball didn't fall/drop as much as
one might predict or calculate. A reference to a 'drop' means the ball fall/drop more than what one might predict or calculate.
